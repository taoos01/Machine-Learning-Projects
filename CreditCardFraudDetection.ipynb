Importing the necessary dependencies

```python
# Dataset: https://www.kaggle.com/datasets/mlg-ulb/creditcardfraud
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# Separating the data for analysis
legit = credit_card_dataset[credit_card_dataset.Class == 0]
fraudaulent = credit_card_dataset[credit_card_dataset.Class == 1]
```

Undersampling

Building a sample dataset containing similar distribution of normal transaction and the fraudaulent transaction

No of fraudaulent transaction -> 492

Randomly taking 492 rows of data from legit transaction

```python
legit_sample =legit.sample(n= 492)
```

Concatinating two dataframes

```python
new_dataframe = pd.concat([legit_sample, fraudaulent], axis= 0) 
```

```python
new_dataframe.head()
```

```python
new_dataframe['Class'].value_counts()
```

Splitting the label and the features

```python
X = new_dataframe.drop(columns= 'Class', axis= 1)
Y = new_dataframe['Class']
```

Train test split

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size= 0.2, stratify= Y, random_state= 2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
print(Y.shape, Y_train.shape, Y_test.shape)
```

Model training

```python
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

Model Evaluation

```python
# Accuracy Score on training data
training_data_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(training_data_prediction, Y_train) * 100
```

```python
# Accuracy on training data is 94.66% which is excellent
print("Accuracy on training data:", round(training_data_accuracy, 2))
```

```python
# Accuracy Score on the test data
test_data_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(test_data_prediction, Y_test) * 100
```

```python
# Accuracy on training data is 91.88% which is again excellent
print("Accuracy on test data:", round(test_data_accuracy, 2))
```
